```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor

# Download and prepare the data
#data_root = "file:///Users/chaitanyanitinkumarranshur/Downloads/API_NY.GDP.PCAP.CD_DS2_en_csv_v2_2702/API_NY.GDP.PCAP.CD_DS2_en_csv_v2_2702.csv"
lifesat = pd.read_csv("/Users/chaitanyanitinkumarranshur/Downloads/gdp-vs-happiness/gdp-vs-happiness.csv")
lifesat = lifesat[["GDP per capita, PPP (constant 2021 international $)", "Cantril ladder score"]].dropna()
X = lifesat[["GDP per capita, PPP (constant 2021 international $)"]].values
y = lifesat[["Cantril ladder score"]].values

# Visualize the data
lifesat.plot(kind='scatter', grid=True,
        x="GDP per capita, PPP (constant 2021 international $)", 
        y="Cantril ladder score")
plt.xlabel("GDP per capita (PPP, 2021 $)")
plt.ylabel("Happiness Score (Cantril ladder)")
plt.title("GDP vs Happiness Score")
plt.show()

# Select a k-nearest neighbors regression model
model = KNeighborsRegressor(n_neighbors=3)

# Train the model
model.fit(X, y)

# Make a prediction for India
X_new = [[9160.063]]  # India's GDP per capita in 2023
print(model.predict(X_new)) # outputs
```
